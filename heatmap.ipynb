# categorical heatmap representing students perfomance.

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
unit = ["physics", "chemistry", "statistics", "complex analysis", "real analysis", "java", "python"]
students= ["Joseph", "Paul", "Timothy", "Loyce", "Lynn", "Simon","Kelly"]
perfomance = np.array([[90, 60, 40, 20, 30, 70, 45],
                     [43, 69, 40, 80, 42, 64, 42],
                     [87, 77, 87, 27, 57, 72, 10],
                     [64, 34, 62, 76, 28, 60, 70],
                     [91, 65, 66, 44, 40, 40, 20],
                     [44, 68, 50, 70, 90, 54, 40],
                     [0.0, 78, 49, 78, 38, 66, 60]])

fig, ax = plt.subplots(figsize=(10,5))
im =ax.imshow(perfomance)

# to show all the ticks
ax.set_xticks(np.arange(len(students)))
ax.set_yticks(np.arange(len(unit)))

# lebelling them with respective entries
ax.set_xticklabels(students)
ax.set_yticklabels(unit)

# aligning the tick labels
plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# loop over data dimensions and create text anotations
for i in range(len(unit)):
    for j in range(len(students)):
        text = ax.text(j, i, perfomance[i, j], ha="center", va="center", color="k")
        ax.set_title("students performance per unit")
        fig.tight_layout()
        plt.show()
        
```
